```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled.describe()
```

```python
pd.DataFrame(data_scaled).describe()
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++')

# fitting the k means algorithm on scaled data
kmeans.fit(data_scaled)
```

```python
kmeans.inertia_
```

```python
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster, init='k-means++')
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)

# converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
# k means using 5 clusters and k-means++ initialization
kmeans = KMeans(n_clusters = 6, init='k-means++')
kmeans.fit(data_scaled)
pred = kmeans.predict(data_scaled)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
data_2d = pca.fit_transform(data_scaled)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the two-dimensional data
plt.figure(figsize=(10, 6))
colors = plt.cm.rainbow(np.linspace(0, 1, len(set(pred))))

for cluster in set(pred):
    plt.scatter(data_2d[pred == cluster, 0], data_2d[pred == cluster, 1],
                color=colors[cluster], label=f'Cluster {cluster}', alpha=0.7)

# Optionally, plot the centroids
centroids_2d = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids_2d[:, 0], centroids_2d[:, 1], s=100, c='black', label='Centroids', marker='X')

plt.title('Clusters Visualization')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```

```python
from sklearn.manifold import TSNE

# Initialize t-SNE
tsne = TSNE(n_components=2, random_state=42)

# Transform the data
data_2d_tsne = tsne.fit_transform(data_scaled)

# Now plot the transformed data with cluster assignments
plt.figure(figsize=(10, 6))
colors = plt.cm.rainbow(np.linspace(0, 1, len(set(pred))))

for cluster in set(pred):
    plt.scatter(data_2d_tsne[pred == cluster, 0], data_2d_tsne[pred == cluster, 1],
                color=colors[cluster], label=f'Cluster {cluster}', alpha=0.7)

plt.title('Clusters Visualization with t-SNE')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.legend()
plt.show()
```
